```
import numpy as np
from matplotlib import pyplot as plt
```

```
t_half = 5.0 * (60*24) # half life in minutes
```

```
step_reduction = 0.5**(1/t_half)
```

```
t = [0]
y = [0]
for i in range(3600*7*10):
  t.append( t[-1]+1 )
  y.append( y[-1]*step_reduction )
  
  if i%3600==0:
    if i>= 3600*7*8:
      dose = 0
    elif i>= 3600*7*5:
      dose = 20
    elif i >= 3600*7*2:
      dose = 40
    else:
      dose = 20
    y[-1] += dose
  
t = np.array(t)
y = np.array(y)
```

```
plt.figure(figsize=(14,8))
plt.plot(t,y)
plt.xticks([w*3600*7 for w in range(11)], range(11))
plt.title("Fluoxetine metabolism")
plt.xlabel("weeks")
plt.ylabel("total body content (mg)")
plt.show()
```
